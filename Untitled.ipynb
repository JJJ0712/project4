```python
from urllib.request import urlopen # 서버 요청/응답 패키지
import bs4
import pandas as pd
```

# 네이버 뉴스 크롤링
- 네이버 언론사별 랭킹뉴스 크롤링
- 매 시간 마다 이전 한 시간 조회수를 기준으로 집계한 랭킹이 노출됨, 1시간 단위로 집계 결과 반영


```python
url = 'https://news.naver.com/main/ranking/popularDay.naver'
html = urlopen(url)
```

```python
bs_obj = bs4.BeautifulSoup(html,'html.parser')
```

```python
print(bs_obj.prettify())
```

# 언론사 리스트

```python
# 언론사 이름 리스트
press = bs_obj.findAll('strong', {'class':'rankingnews_name'})
```

```python
press_list = []
for i in press:
    press_list.append(i.text)
```

```python
# 언론사명, 랭킹뉴스 1~5위 타이틀, url df 생성
rank = pd.DataFrame(index = press_list)
```
