```python
import pandas as pd
```

```python
source = pd.read_csv('ranking_section.csv')
```

```python
source
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환

count_vect = CountVectorizer(min_df= 0, ngram_range=(1,2))# ngram_range=(1,2)1개 또는 2개 단어를 묶어서 벡터화
count_vect.fit(source['content'])
```

```python
count_vect.get_feature_names()
```
